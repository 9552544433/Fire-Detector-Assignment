```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
#####function to load and classify the image
def classify_and_display_image(model, img_path, target_size=(150, 150)):
    
    img = image.load_img(img_path, target_size=target_size)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0  

    # predict using the model
    prediction = model.predict(img_array)

    
    label = "No Fire" if prediction[0][0] > 0.5 else "Fire"

    # the image and its classification
    plt.imshow(img)
    plt.title(f'Classified as: {label}')
    plt.axis('off')
    plt.show()

model = tf.keras.models.load_model('best_fire_detection_model.keras')
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss='binary_crossentropy',
              metrics=['accuracy'])
classify_and_display_image(model,r"C:\Users\PC\Desktop\6 projects\gettyimages-1447708518-612x612.jpg")
```
